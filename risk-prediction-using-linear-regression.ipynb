```python
# problem statement ---> What is the risk for given data?
# speed=160 km/h, car_age=10 years, experience= 5 years

"""
Y= MX+C
risk = M1*speed + M2*car_age + M3*experience + intercept

"""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
df=pd.read_csv("/kaggle/input/datasets/rkhademul19/car-data/car data.csv")
df
```

```python
# fill up the null value using mean or median

exp_fit=df.experience.median()
exp_fit
```

```python
# fillna() funtion to fill the null data

df.experience= df.experience.fillna(exp_fit)
df
```

```python
# Apply Linear Regression
from sklearn.linear_model import LinearRegression
reg= LinearRegression()
```

```python
# model train
reg.fit(df[['speed','car_age','experience']],df.risk)
```

```python
#  risk of the given data

reg.predict([[160,10,5]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
